```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,4,6,9,16,25]
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot ')
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [11,24,36,56,61,95]
plt.plot(x,y,color='red',linestyle = '-.',marker = 'o',linewidth ='1.5',markersize = 7)

plt.grid(True)
```
