```python
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f'Hey, amount is deposited, new balance is {self.balance}')
    def withdrawal(self,amount):
        if amount>self.balance:
            print("Hey, Insufficient Funds")
        else:
            self.balance-=amount
            print(f"hey the {amount} has been withdrawn, the new balance is {self.balance}")
    def get_balance(self):
        return self.balance
    
# creata an object
account=BankAccount("pratham",5000)
print(account.balance)
account.withdrawal(4000)
account.deposit(10)
account.withdrawal(5000)
print(account.get_balance())
```
