```python
def conver_temp(temp,unit):
    if unit == 'C':
        return temp * 9/5 + 32
    elif unit == 'F':
        return (temp - 32)*5/9
    else:
        return None
    
conver_temp(33 ,"C")
```

```python
def is_strong_password(password):
    """This fun checks is the password is strong or not."""
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
```

```python
def calculate_totalcost(cart):
    total_cost=0
    for item in cart:
        total_cost+= item["price"]*item['quantity']
    return total_cost

cart=[{'name':'apple','price':0.9,'quantity':4},
    {'name':'banana','price':0.5,'quantity':6},
    {'name':'orange','price':0.7,'quantity':5}]

total_cost=calculate_totalcost(cart)
print(total_cost)
```
