```python
# Base Class
class Animal:
    def speak(self):
        return "Sound of the Animal"

# Derived Class
class Dog(Animal):
    def speak(self):
        return "whoof !"

class Cat(Animal):
    def speak(self):
        return "Meow !"

dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())
```
