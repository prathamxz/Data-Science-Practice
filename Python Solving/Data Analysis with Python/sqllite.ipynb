```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
```

```python
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute(
'''
Create table if not exists employees(
    id Integer Primary key,
    name Text not null,
    age Integer , 
    department text
)'''
)
connection.commit
```

```python
# Insert the data
cursor.execute('''
Insert Into employees(name, age, department)
               values('Pratham', 23, 'Scientist')''')
cursor.execute('''
Insert Into employees(name, age, department)
               values('Bob', 21, 'Data Engineer')''')
cursor.execute('''
Insert Into employees(name, age, department)
               values('Jolly', 32, 'LLB')''')
cursor.execute('''
Insert Into employees(name, age, department)
               values('Isha', 43, 'Engineer')''')
```
